```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
import time
import mediapipe as mp
```

# Ključne tačke - ruke, telo, glava

```python
mp_holistic = mp.solutions.holistic #  model za identifikaciju lica, tela i saka
mp_drawing = mp.solutions.drawing_utils # utilities 
```

```python
#pomocna funckija, uzima sliku (frame, i to je u BGR blue green red redosledu boja u array-u) i model 
def mediapipe_detection(image, model):    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # mediapipe ocekuje RGB, image je u BGR, pa je moramo konvertovati
    image.flags.writeable = False                  # ne mogu da se prave slucajne izmene nad slikom
    results = model.process(image)                 # predikcija, analizira rgb sliku i vraca rezultate detekcije lica itd
    image.flags.writeable = True                   # vracanje nazad podesavanja za sliku, kako bismo iscrtali tacke
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # konverzija slike nazad u BGR kako bi opet OpenCv mogao kako treba ad ocita
    return image, results
#vraca na krau samu sliku ali media pipe objekat (results) sa podacima sa kojima ce posle crtati
```

```python
def draw_styled_landmarks(image, results):  # parametre koje funkcija prima: slika, rezultat -tackice, koncekcije 
    
    #lice
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS, 
                             mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1), 
                             mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
                             ) 
    # poza
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                             ) 
    # leva ruka
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                             ) 
    # desna ruka 
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                             ) 
```

```python
#proba - da se ispitanik namesti
cap = cv2.VideoCapture(0)

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()
        image, results = mediapipe_detection(frame, holistic)
        print(results)
        draw_styled_landmarks(image, results)
        cv2.imshow('OpenCV Feed', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```

```python
def extract_keypoints(results):   # za svaku tacku pamte se koordinate : x,y,z i vidljivost (za pose).. Ukoliko nema ispisi 0
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33*4)
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([pose, face, lh, rh])   
```

# Pravljenje kolekcija za trening

```python
import os
```

```python
DATA_PATH = os.path.join('Master_Data') 
actions = np.array(['Ja','Mogu','Riba','Sve','Momak','Devojka','Opasno','Kada','Zasto','Hvala'])
no_sequences = 20
sequence_length = 35
gender = 'man'  # musko ili zensko, ime ucestnika -> menja se svaki put kad se promeni ucesnik
person = 'person4'
```

```python
for action in actions: 
    for sequence in range(no_sequences):
        try: 
            os.makedirs(os.path.join(DATA_PATH,gender, person, action, str(sequence)))
        except:
            pass
```

```python
# Kamera se opet pali da bi se napravio  set podataka
# Za svaku reč se pravi 20 snimaka, a za svaki snimak se pravi po 35 framova, svaki frame se čuva kao kolekcija

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Camera not accessible.")
else:
    with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
        
        actions = np.array(['Hvala'])
        for action in actions:
            time.sleep(1) 
            for sequence in range(no_sequences):
                print(f"Get ready for gesture '{action}', video {sequence+1}/{no_sequences}")
                #time.sleep(1) 

                for frame_num in range(sequence_length):
                    ret, frame = cap.read()
                    if not ret:
                        print("Error: Failed to read frame.")
                        break
                        
                    image, results = mediapipe_detection(frame, holistic)
                    draw_styled_landmarks(image, results)
                    
                    if frame_num == 0: 
                        cv2.putText(image, 'SPREMI SE! REC {} !'.format(action), (120,200), 
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255, 0), 4, cv2.LINE_AA)
                        cv2.putText(image, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12), 
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
                        cv2.imshow('OpenCV Feed', image)
                        cv2.waitKey(2000)
                    else: 
                        cv2.putText(image, 'Collecting frames for {} Video Number {}'.format(action, sequence), (15,12), 
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
                        cv2.imshow('OpenCV Feed', image)
                    
                    keypoints = extract_keypoints(results)
                    npy_path = os.path.join(DATA_PATH,gender,person,action, str(sequence), str(frame_num))
                    np.save(npy_path, keypoints)

                    if cv2.waitKey(10) & 0xFF == ord('q'):
                        break

    cap.release()
    cv2.destroyAllWindows()
```
